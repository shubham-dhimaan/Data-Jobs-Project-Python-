```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Concat Job Postings (2.8.1)

```python
job_titles = df['job_title_short'].unique()

# write this with dict comprehension
dict_titles = {job_title: df[df['job_title_short'] == job_title] for job_title in job_titles}

type(dict_titles) #created a dictionay or dataframes by month
```

```python
dict_titles['Data Analyst']
```

```python
two_roles = [dict_titles['Data Analyst'], dict_titles['Data Engineer']]

df_conact_1 = pd.concat(two_roles, ignore_index= True) #Both Full-time and internship jobs joined

df_conact_1
```

## 🟨 Concat Job Types (2.8.3)

```python
# make data frame for each job_schedule_type using a loop
job_types = df['job_schedule_type'].unique()

# write this with dict comprehension
dict_types = {job_type: df[df['job_schedule_type'] == job_type] for job_type in job_types}

type(dict_types) #created a dictionay or dataframes by month
```

```python
dict_types['Full-time']

```

```python

two_types = [dict_types['Full-time'], dict_types['Internship']]

df_conact = pd.concat(two_types, ignore_index= True) #Both Full-time and internship jobs joined

df_conact
```
