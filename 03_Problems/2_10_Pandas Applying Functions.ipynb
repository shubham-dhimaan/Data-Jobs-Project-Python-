```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Convert Date to String (2.10.1)

```python
df['job_posted_weekday'] = df['job_posted_date'].apply(lambda x: x.strftime('%a')) #creating new col for weekday name

df
```

---

## 🟨 Days Since Posted (2.10.2)

```python
df['days_since_posted'] = df['job_posted_date'].apply(lambda x: (pd.Timestamp.now() - x)).dt.days

# applied lambda function on each row

# to subtract current date time from posted date time 

# then accessed days using dt

df
```

## 🟥 Salary Category (2.10.3)

```python
"""
Salary categories based on quartiles (percentiles) of salary_year_avg:

    <= 25th percentile             -> Low Salary
    > 25th and <= 50th percentile  -> Average Salary
    > 50th and <= 75th percentile  -> Above Average Salary
    > 75th percentile              -> High Salary

    """

# Step 1: Remove NaN values in salary and getting quartiles values 

df_salary = df[df['salary_year_avg'].notna()].copy()

quartiles = df_salary['salary_year_avg'].quantile([0.25, 0.50, 0.75]).values.tolist() # Storing quartile values for salary_year_avg cols in a list

# Step 2: Writing a function using dynamic quartile values to categorize salary

def SalaryCategory(salary):
    if salary <= quartiles[0]:
        salary_category = 'Low Salary'
    elif salary > quartiles[0] and salary  <= quartiles[1]:
        salary_category = 'Average Salary'
    elif salary > quartiles[1] and salary  <= quartiles[2]:
        salary_category = 'Above Average Salary'
    else:
        salary_category = 'High Salary'
    return salary_category


df_salary['yearly_salary_category'] = df_salary['salary_year_avg'].apply(SalaryCategory)

df_salary
```
