```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanupy

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
def inflation(salary):
    return salary * 1.03

df['salary_year_inflated'] = df['salary_year_avg'].apply(inflation)

df[pd.notna(df['salary_year_avg'])][['salary_year_avg', 'salary_year_inflated']] #Displaying only where fucntion is applied
```

```python
df['salary_year_inflated'] = df['salary_year_avg'].apply(lambda salary: salary * 1.03) # Written simply with lambda

df[pd.notna(df['salary_year_avg'])][['salary_year_avg', 'salary_year_inflated']]
```
