```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanupy

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')
```

```python
# make data frame for each month using a loop
months = df['job_posted_month'].unique()

# write this with dict comprehension
dict_months = {month: df[df['job_posted_month'] == month] for month in months}

# long way
# dict_months = {}
# for month in months:
#     dict_months[month] = df[df['job_posted_month'] == month]

type(dict_months) #created a dictionay or dataframes by month
```

```python
dict_months['Jan'] #calling the dictonary with key 'Jan'annd value is the df for month january that's returned
```

```python
# Concatinating for Quater 2

quarter_2 =  [dict_months['Apr'], dict_months['May'], dict_months['Jun']]

df_q2 = pd.concat(quarter_2, ignore_index= True)  #ignore index will discard old index and create a new one

df_q2
```
