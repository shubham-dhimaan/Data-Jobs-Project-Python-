```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanupy

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.sample(3)
```

```python
df.index.dtype #int64

df.index.name # No name for the idex is set (no title over index column)
```

```python
df.index.name = 'job_index'

df.sample(3) # name is set
```

```python
salary_stats = df.pivot_table(values= 'salary_year_avg', index= 'job_title_short', aggfunc= ['median', 'mean', 'max'])

salary_stats #index is already created with the column name
```

```python
salary_stats.index.name
```

```python
df_in = df[df['job_country'] == 'India']

df_in.reset_index(inplace= True) # old index is set aside (kept) and new index takes palce

df_in.head(5) # The new index again won't have a name, job_index now becomes a column
```

```python
salary_stats.reset_index(inplace= True) #here job_title_short is no longer the index

salary_stats
```

```python
df_in.set_index('job_index', inplace=True)  #can be only run ONCE; discard old index and used mention col(s) as new index

df_in #main difference between reset and set index is that set index discards the old index while reset one adds it as a col
```

```python
salary_stats.set_index('job_title_short', inplace= True)

salary_stats
```

```python
salary_stats.sort_values(by=('median', 'salary_year_avg'), ascending=False, inplace= True)

salary_stats #changing sort order for demonstration
```

```python
salary_stats.sort_index() #sorts back the dataframe using the index alphabetically, useful when index doesn't have name
```
