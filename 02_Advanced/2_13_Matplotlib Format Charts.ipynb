```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

## 🟩 Subplots with Bar Charts (2.13.1)

---

## 🟨 Subplots with Job Titles by Schedule (2.13.2)

```python
# Code Needed
schedules = ['Full-time', 'Part-time', 'Contractor', 'Internship']
job_titles = ['Data Analyst', 'Senior Data Analyst', 'Data Engineer', 'Machine Learning Engineer']

df_schedule_pivot = df.pivot_table(index= 'job_schedule_type', columns= 'job_title_short', aggfunc='size', fill_value=0)

df_plot = df_schedule_pivot.loc[schedules, job_titles]

fig, ax = plt.subplots(len(schedules), 1)

for i, schedule in enumerate(schedules):
    ax[i].barh(df_plot.loc[schedule].index, df_plot.loc[schedule].values)
    ax[i].set_title(schedule)

fig.tight_layout()
```
